# Setup

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/customer_shopping_behavior.csv')
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isna().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isna().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount_usd'})
df.columns
```

# Feature engineering

```python
# Creating column age_group
labels = ['Young Adults', 'Adults', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
df[['age','age_group']].head()
```

```python
df.age_group.value_counts()
```

```python
df.frequency_of_purchases.unique()
```

```python
# Creating column purchase_frequency_days
mapping = {
     'Fortnightly': 14,
     'Weekly': 7,
     'Annually': 365,
     'Quarterly': 90,
     'Bi-Weekly': 14,
     'Monthly': 30,
     'Every 3 Months': 90
}

df['purchase_frequency_days'] = df.frequency_of_purchases.map(mapping)
```
